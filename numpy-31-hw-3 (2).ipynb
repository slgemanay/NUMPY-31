```python
import pandas as pd
```

```python
data = pd.read_csv('ml-latest-small/ratings.csv')
data.head()
```

```python
def rating_detection(rating):
    if rating <= 2.0:
        return 'Низкий рейтинг'
    if 2.0 < rating <= 4.0:
        return 'Средний рейтинг'
    if rating == 4.5 or 5.0:
        return 'Высокий рейтинг'
```

```python
data['rating_assessment'] = data['rating'].apply(rating_detection)
data.head()
```

```python
import numpy as np
keywords = pd.read_csv('../input/numpy31hw3/keywords.csv')
keywords.head()
```

```python
geo_data = {
    
'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def geo_classifier(keyword):
    for region, city_list in geo_data.items():
        for city in city_list:
            if city in keywords['keyword']:
                keywords['Region'] = region
    else:
        keywords['Region'] = 'undefined'
```

```python
geo_classifier(keywords)
```
